# **Part 2: Data transformation**

In this part of the code, we are preparing the dataset (german_credit) by enhancing it with detailed attribute columns, adjusting the target variable, and subsequently saving the modified dataset to a CSV file (german_credit.csv).

```python
# Combine features and targets into a single DataFrame
german_credit = pd.concat([X, y], axis=1)
```

```python
# Check
german_credit.head(5)
```

The overall information about the credit portfolio provided for developing a scoring model is obtained by combining previously loaded and processed data. In this merged database, each column (21 in total) corresponds to a single feature (20 Attributes and 1 class), with the respective cell values reflecting its possible values.

The data in the provided database does not meet the requirements for use in model development. Consequently, using the obtained feature values, the number of features increases, and the data is converted into a binary representation.

Each column in the reprocessed database reflects a specific quantitative feature. Additionally, the number of attributes has increased because those that had different values have been split into multiple attributes according to the number of inherent values, and the values in the cells of each column have been converted into a binary system.

```python
# Columns for the detailed attributes in german_credit:
for col_name in subname_list:
    german_credit[col_name] = german_credit.apply(lambda row: 1 if col_name in row.values else 0, axis=1)

# Getting all values from the added columns
values_to_remove = subname_list

# Lambda function to remove columns containing values from values_to_remove
remove_columns = lambda df, values: df.drop(columns=[col for col in df.columns if df[col].isin(values).any()])

# Applying the lambda function to german_credit
german_credit = remove_columns(german_credit, values_to_remove)

german_credit['class'] = german_credit['class'].apply(lambda x: 0 if x == 1 else 1)

# Data for future analysis
german_credit
```

# **Part 3: Models for analysis**

**3.1. Models**

These models are chosen based on their respective strengths in handling different aspects of data, complexity of relationships, and performance metrics relevant to credit scoring tasks. Each model may excel in different scenarios, and ensemble methods like boosting and random forests often complement traditional linear models like logistic regression.
- **Logistic Regression**: Often used for its simplicity and interpretability in binary classification tasks like credit scoring. It models the probability of a binary outcome based on input features.

- **Random Forest**: Effective for handling complex relationships in data and avoiding overfitting. It builds multiple decision trees and combines their predictions to improve accuracy.

- **Gradient Boosting**: Builds an ensemble of trees sequentially, where each subsequent tree corrects the errors made by the previous one. It's powerful for predictive accuracy.

- **XGBoost**: An optimized implementation of gradient boosting known for its speed and performance. It supports regularization techniques to prevent overfitting and is widely used in competitions.

- **SVM (Support Vector Machine)**: Effective for high-dimensional data and finding complex decision boundaries. It maps input data to a higher-dimensional space and finds the optimal separating hyperplane.

- **KNN (K-Nearest Neighbors)**: Classifies data points based on the majority class among their nearest neighbors. It's intuitive and non-parametric but can be computationally expensive with large datasets.

- **Naive Bayes**: Assumes independence between features and calculates the probability of each class given the input features. It's simple, fast, and works well with categorical data.

- **LightGBM**: A gradient boosting framework that uses tree-based learning algorithms. It's known for its speed and efficiency, particularly with large datasets.

- **CatBoost**: Another gradient boosting library that handles categorical variables well without the need for extensive preprocessing. It's robust and performs well out-of-the-box.

- **Neural Network (MLPClassifier)**: Learns complex patterns in data through multiple layers of interconnected nodes. It's powerful for capturing non-linear relationships but requires tuning and careful regularization.

**3.2. Features for models based on correlation**

In this part of the code, we determine the correlation between features and the target. Additionally, we create two lists of features: **highly_correlated_features** - the first list includes all features with correlations above the mean, and **correlated_features** - the second list includes features with correlations not lower than the mean of the entire list minus the standard deviation.

```python
# correlations
corr_matrix = german_credit.corr()

# correlations matrix
print(corr_matrix.shape)

# Correlations with the target variable (class)
correlations = german_credit.corr()['class'].drop('class')

# Sorting by the absolute value of correlation with the target variable (class)
correlations = correlations.abs().sort_values(ascending=False)

# average correlation with the target variable (class)
mean_correlation = np.mean(correlations)

# setting the lower threshold of applicability for variables
std_deviation = np.std(correlations)
lower_threshold = mean_correlation - std_deviation

# features with correlation with the target variable (class) > mean
highly_correlated_features = correlations[abs(correlations) > mean_correlation]

# number of the features with correlation with the target variable (class) > mean
len(highly_correlated_features)

# information about the features with correlation with the target variable (class) > mean
print(f'There are', len(highly_correlated_features),'highly correlated with "class" features. They have correlations > than',mean_correlation)

# features with correlation with the target variable (class) > lower_threshold
correlated_features = correlations[abs(correlations) > lower_threshold]

# number of the features with correlation with the target variable (class) > lower_threshold
len(correlated_features)

# information about the features with correlation with the target variable (class) > lower_threshold
print(f'There are', len(correlated_features),'correlated with "class" features. They have correlations > than',lower_threshold)
```

**3.3. Models Testin**

At the initial stage of testing, we will use the list of features called **correlated_features** to evaluate the performance and accuracy of models and their cost when using the full list of features. The cost of the model is determined according to the cost matrix provided on the website **https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data**, where it is defined that the cost of misclassifying a good client as bad is 1, and the cost of misclassifying a bad client as good is 5.

```python
# set error costs
cost_fp = 1
cost_fn = 5
```

We conduct testing on the full original dataset, then normalize the data and evaluate the models using the normalized data. Additionally, we assess the performance of the models using **SMOTE** and **GridSearch**. The research results will be stored in a database **full_models_evaluation_correlated_features** for further use in determining the optimal model with the optimal number of variables.
To perform the evaluation we use **two functions**:
 - **evaluate_models** - evaluates multiple models based on various metrics including accuracy, precision, recall, F1-score, and cost, using confusion matrix analysis and class weights
 - **evaluate_models_gr** - evaluates multiple models using GridSearchCV for parameter tuning (if specified), and computes evaluation metrics including accuracy, precision, recall, F1-score, and cost.

Both functions allow you to see the values of **Accuracy** and **Cost**, as well as the **Confusion Matrix** for each tested model during execution.

```python
# Testing using the correlated_features
models = start_models

features = german_credit[correlated_features.index]
target = german_credit['class']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

fuetures_type = "correlated features"
results_before_correlated_features = evaluate_models(X_train, X_test, y_train, y_test, "Before Processing", fuetures_type, models, cost_fp, cost_fn)

df_results_before_correlated_features = pd.DataFrame(results_before_correlated_features).sort_values(by='Cost', ascending=True)
#df_results_before_correlated_features

# Normalization
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)
results_after_norm_correlated_features = evaluate_models(X_train_norm, X_test_norm, y_train, y_test, "After Normalization", "correlated features",models, cost_fp, cost_fn)

df_results_before_correlated_features_norm = pd.DataFrame(results_after_norm_correlated_features).sort_values(by='Cost', ascending=True)
#df_results_before_correlated_features_norm

# SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_norm, y_train)
results_correlated_features_norm_after_smote = evaluate_models(X_train_resampled, X_test_norm, y_train_resampled, y_test, "After SMOTE","correlated features", models, cost_fp, cost_fn)

df_results_correlated_features_norm_after_smote = pd.DataFrame(results_correlated_features_norm_after_smote).sort_values(by='Cost', ascending=True)
#df_results_correlated_features_norm_after_smote

features_type = "correlated features"
results_correlated_features_after_gridsearch = evaluate_models_gr(X_train_resampled, X_test_norm, y_train_resampled, y_test, "After GridSearch", features_type, models, cost_fp, cost_fn)
df_results_correlated_features_after_gridsearch = pd.DataFrame(results_correlated_features_after_gridsearch).sort_values(by='Cost', ascending=True)
#df_results_correlated_features_after_gridsearch

full_models_evaluation_correlated_features = pd.concat([df_results_before_correlated_features, df_results_before_correlated_features_norm, df_results_correlated_features_norm_after_smote, df_results_correlated_features_after_gridsearch]).sort_values(by='Cost', ascending=True)
full_models_evaluation_correlated_features.sort_values(by='Cost', ascending=True)
full_models_evaluation_correlated_features.reset_index(drop=True, inplace=True)

display(HTML('<b>Ranking of models by Cost (prepared with correlated features)</b>'))
display(HTML(full_models_evaluation_correlated_features.head().to_html()))
```

**3.5. Defining features for the optimal model (ANOVA F-test features selection)**

Since all the features in our case are numerical, we will use an ANOVA F-test to determine the optimal variables for the model.

The goal of this code is to demonstrate a structured approach to feature selection and model evaluation using a Gradient Boosting Classifier. Feature selection is performed using ANOVA F-test to select the top 10 best features based on their importance. The selected features are then used to train the model and evaluate its performance using cross-validation. Additionally, GridSearchCV is employed to optimize the model's hyperparameters, aiming to improve predictive accuracy.

```python
y_true = german_credit['class']
# Load the saved model
model_filename = 'gradient_boosting_model.pkl'
with open(model_filename, 'rb') as file:
    loaded_model = pickle.load(file)

# Load the selected features
features_filename = 'selected_features.pkl'
with open(features_filename, 'rb') as file:
    selected_features_corr_value = pickle.load(file)

# Load the scaler
scaler_filename = 'standard_scaler.pkl'
with open(scaler_filename, 'rb') as file:
    scaler = pickle.load(file)

# Select only the relevant features
X_new = german_credit[selected_features_corr_value]

# Scale the new data using the loaded scaler
X_new_scaled = scaler.transform(X_new)

# Make predictions using the loaded model
predictions = loaded_model.predict(X_new_scaled)
predictions_proba_class1 = loaded_model.predict_proba(X_new_scaled)[:, 1]
```

```python
# Calculate Precision-Recall curve and average precision
precision, recall, thresholds_pr = precision_recall_curve(y_true, predictions_proba_class1)
average_precision = average_precision_score(y_true, predictions_proba_class1)

# Plot Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.step(recall, precision, color='blue', where='post', label=f'AP = {average_precision:.2f}')
plt.fill_between(recall, precision, step='post', alpha=0.2, color='blue')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curve')
plt.legend(loc="lower right")
plt.show()

# Get class predictions
predicted_class = model.predict(X_new_scaled)
```

```python
# Plot confusion matrix
cm = confusion_matrix(y_true, predicted_class)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=['Class 0', 'Class 1'],
            yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Get ROC curve data
fpr, tpr, thresholds_roc = roc_curve(y_true, predictions_proba_class1)
roc_auc = roc_auc_score(y_true, predictions_proba_class1) 

plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```

```python
# Build confusion matrix
cm = confusion_matrix(y_true, predicted_class)

# Generate classification report dictionary
class_report_dict = classification_report(y_true, predicted_class, output_dict=True)

# Define model description dictionary
model_description = {
    "Model": "Gradient Boosting Classifier",
    "Algorithm": "Gradient Boosting",
    "Features": selected_features_corr_value,
    "Classes": ['Class 0', 'Class 1'],
    "Parameters": model.get_params(),
    "Accuracy": model.score(X_new_scaled, y_true),
    "Confusion Matrix": cm.tolist(),
    "Classification Report": class_report_dict,
    "Feature Importances": model.feature_importances_.tolist(),
    "Number of Estimators": model.n_estimators,
    "Learning Rate": model.learning_rate
}

# Optionally, include other relevant metrics specific to your analysis
```

```python
# Print model description
print("\nScoring Model Description:")
for key, value in model_description.items():
    print(f"{key}: {value}")
```

```python
# Check if the file exists and remove it if it does
description_file = 'credit_scoring_model_description.txt'
if os.path.exists(description_file):
    os.remove(description_file)

# Save model description to file
with open(description_file, 'w') as f:
    f.write("Scoring Model Description:\n")
    for key, value in model_description.items():
        f.write(f"{key}: {value}\n")
    f.write("\nExamples of Model Predictions:\n")
    for i, (true_class, pred_class, prob) in enumerate(zip(y_true, predicted_class, predictions_proba_class1)):
        f.write(f"Example {i + 1}: True class: {'Class 1' if true_class == 1 else 'Class 0'}, Predicted class: {'Class 1' if pred_class == 1 else 'Class 0'}, Probability: {prob:.2f}\n")

print(f"\nScoring model and description saved to files: 'credit_scoring_model.pkl', '{description_file}'")
```

- Accuracy:

Class 1 Precision (1): Precision for Class 1 is 81.7%, indicating that when the model predicts an instance as Class 1, it is correct 81.7% of the time.
Recall:

Class 0 Recall (0): Recall measures how many of the actual Class 0 instances the model correctly predicts as Class 0. It is 49.7%, indicating that the model identifies 49.7% of all Class 0 instances.
Class 1 Recall (1): Recall for Class 1 is 96.1%, suggesting that the model correctly identifies 96.1% of all Class 1 instances.
F1-score:

Class 0 F1-score (0): The harmonic mean of precision and recall for Class 0 is 62.6%. It balances precision and recall into a single metric.
Class 1 F1-score (1): The F1-score for Class 1 is 88.3%, indicating a balanced performance between precision and recall for Class 1.
Support: The number of occurrences of each class in the dataset. In this case, there are 300 instances of Class 0 and 700 instances of Class 1.

- Interpretation:

The model shows strong performance in correctly predicting Class 1 instances (high recall and F1-score), which might be critical depending on the application (e.g., identifying positive cases in medical diagnostics).
However, the model's performance is weaker for Class 0 in terms of recall, indicating it misses nearly half of the actual instances of Class 0.
Precision for both classes is relatively high, suggesting that when the model predicts a class, it tends to be correct a significant portion of the time.
The F1-scores provide a balanced view of the model's ability to achieve both high precision and recall, with Class 1 performing particularly well compared to Class 0.
